```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('progress.csv', header = 0)
pred_anno = []
pred_correct = []
final_anno = []
false_positives_PR = []
true_positives_PR = []
x_pos = []
for index, row in df.iterrows():
    if row['Round Used'] > 0:
        pred_anno.append(row['Anno Pred'])
        pred_correct.append(row['Correct Pred'])
        final_anno.append(row['Total Anno'])
for i in range(0,len(pred_anno),2):
    x_pos.append(str(int(i-i/2)))
    false_positives_PR.append(((pred_anno[i] + pred_anno[i+1]) - (pred_correct[i] + pred_correct[i+1]))/(final_anno[i] + final_anno[i+1]))
    true_positives_PR.append((pred_correct[i] + pred_correct[i+1])/(final_anno[i] + final_anno[i+1]))
plt.bar(x_pos, true_positives_PR)
plt.ylabel('Percent True Postives')
plt.xlabel('Stage')
plt.title("True Positive Annotations Over Time")
plt.ylim(0.5,1.0)
plt.show()
plt.bar(x_pos, false_positives_PR)
plt.ylabel('Percent False Postives')
plt.xlabel('Stage')
plt.title("False Positive Annotations Over Time")
plt.show()
true_pos = [0.36,0.61, 0.65, 0.82, 0.77, 0.88]
stages = ['Stage 1', 'Stage 2', 'Stage 3', 'Stage 4', 'Stage 5', 'Stage 6']
plt.bar(stages, true_pos)
plt.ylabel('Percent True Postives')
plt.xlabel('Stages in Auto-Annotating Labeling')
plt.title("True Positive Annotations Over Auto-Annotating Stages")
plt.show()
```
